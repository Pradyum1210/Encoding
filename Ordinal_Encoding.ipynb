‚úÖ Definition of Encoding

Encoding is the process of converting non-numeric data (text or categories) into numeric values so that machine learning models can understand and process the data.

üìå ML algorithms work only with numbers, so encoding is necessary.

‚úÖ Types of Data

There are two main types of data:

1Ô∏è‚É£ Numerical Data
2Ô∏è‚É£ Categorical Data                            

üîµ 1Ô∏è‚É£ Numerical Data

Numerical data consists of numbers that have mathematical meaning.

Types of Numerical Data:

Discrete ‚Üí Countable numbers
üëâ Example: number of students, number of cars

Continuous ‚Üí Measurable values
üëâ Example: height, weight, temperature

üìå Numerical data usually does not require encoding.

üü¢ 2Ô∏è‚É£ Categorical Data

Categorical data represents labels or categories instead of numbers.

Example:

Gender ‚Üí Male, Female

City ‚Üí Pune, Mumbai, Delhi

Education ‚Üí School, College, PG

üìå Categorical data must be encoded.

üî∏ Types of Categorical Data

2Ô∏è‚É£(a) Nominal Data

Nominal data has NO specific order or ranking.

Examples:

Gender ‚Üí Male, Female

Color ‚Üí Red, Blue, Green

City ‚Üí Pune, Mumbai, Delhi

üìå Order does not matter

Encoding used:

One-Hot Encoding

Label Encoding (sometimes)

2Ô∏è‚É£(b) Ordinal Data

Ordinal data HAS a meaningful order or ranking.

Examples:

Education ‚Üí School < College < PG

Rating ‚Üí Low < Medium < High

Customer feedback ‚Üí Poor < Average < Good < Excellent

üìå Order matters

Encoding used:

Ordinal Encoding

Label Encoding with order

Poor      ‚Üí 1

Average   ‚Üí 2

Good      ‚Üí 3

Excellent ‚Üí 4

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/customer.csv')
df.head()
```

```python
df = df.drop(columns = ['age','gender'])
df.head()
```

```python
x = df[['review',	'education']]

y = df['purchased']
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.3,random_state = 12)
```

**Ordinal Encoding**

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories = [['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(x_train)
```

```python
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
```

```python
oe.categories_
```

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
