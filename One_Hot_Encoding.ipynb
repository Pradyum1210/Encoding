```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

```python
df['owner'].value_counts()
```

**1. One Hot Encoding using Pandas**

```python
pd.get_dummies(df,columns = ['fuel','owner'])
```

**2.  K-1 OneHotEncoding**   
It drop first column and k is the original number of columns

```python
pd.get_dummies(df,columns = ['fuel','owner'],drop_first = True)
```

**3 . OneHotEncoding using Sklearn**

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split

x = df[['brand', 'km_driven', 'fuel', 'owner']]

y = df['selling_price']

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2,random_state = 2)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop = 'first' , sparse_output = False , dtype = np.int32)
```

```python
x_train_new = ohe.fit_transform(x_train[['fuel','owner']])
```

```python
x_test_new = ohe.fit_transform(x_test[['fuel','owner']])
```

```python
x_train.shape , x_train_new.shape , x_test.shape , x_test_new.shape
```

```python
# Combine original numerical columns (brand, km_driven)
# with newly encoded features (x_train_new) column-wise
# np.hstack() stacks arrays horizontally (side by side)
# .values converts DataFrame columns into NumPy arrays


np.hstack((x_train[['brand','km_driven']].values , x_train_new))
```

**4. OneHotEncoding with top categories**

```python
counts = df['brand'].value_counts()
counts
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl,'uncommon'))
```
